### Handling Null Values

#### Size

```python
df['Size'].unique()
```

```python
df.groupby('Category')['Size'].unique()
```

```python
df.groupby('Item Purchased')['Size'].unique()
```

```python
#replacce null values in electronics,accesories,footwea
df.loc[df['Category']=='Electronics','Size'] = 'Not Applicable'
df.loc[df['Category']=='Accessories','Size'] = 'Not Applicable'
df.loc[df['Category']=='Footwear','Size'] = 'Not Applicable'
```

```python
df.loc[(df['Category'] == 'Clothing')&(df['Size'].isnull()),'Size'] = df[df['Category'] == 'Clothing']['Size'].mode()[0]
```

```python
df.groupby('Category')['Size'].unique()
```

```python
df.isnull().sum()
```

#### Review Rating

```python
# mean of review_ratings
df['Review Rating'].mean()
```

```python
# median of review_ratings
df['Review Rating'].median()
```

```python
# mode of review_ratings
df['Review Rating'].mode()
```

```python
df.groupby('Item Purchased')['Review Rating'].mean()
```

```python
#filling null values of review rating with the mean of each item purchased 
df['Review Rating'] = df['Review Rating'].fillna(df.groupby('Item Purchased')['Review Rating'].transform('mean'))
```

```python
df.isnull().sum()
```

```python
df.head()
```

#### Previous Purchase

```python
#filling previous purchases with Zero(0)
df['Previous Purchases'] = df['Previous Purchases'].fillna(0)
```

```python
df.isnull().sum()
```

#### Purchase Amount

```python
#filling null values of purchase amounts with mean of each item purchased 
df['Purchase Amount (USD)'] = df['Purchase Amount (USD)'].fillna(df.groupby('Item Purchased')['Purchase Amount (USD)'].transform('mean'))
```

```python
df.isnull().sum()
```

#### Renaming the Header

```python
df.columns = df.columns.str.replace(' ','_')
df.columns = df.columns.str.lower()

```

```python
from sqlalchemy import create_engine
import pandas as pd
# Replace these values with your actual pgAdmin/PostgreSQL details
db_username = 'postgres'       # Default username is usually 'postgres'
db_password = 'Password'  # The password you set during PostgreSQL installation
db_host = 'localhost'          # Usually 'localhost' for local server
db_port = '5432'               # Default port
db_name = 'data_analyst_project' # The specific database you want to connect to (e.g., 'customer_analysis')

# This is the name of the table you want to create in pgAdmin
TABLE_NAME = 'customer_behavior_analysis'
DB_NAME = 'data_analyst_project'
# Construct the connection string
# Format: postgresql+psycopg2://user:password@host:port/database
connection_string = f"postgresql+psycopg2://{db_username}:{db_password}@{db_host}:{db_port}/{db_name}"

# Create the engine
try:
    engine = create_engine(connection_string)
    print("✅ Successfully connected to PostgreSQL!")
    
    # Test connection by reading a system table (optional)
    df_test = pd.read_sql_query("SELECT version()", engine)
    print("Connection Test Result:")
    print(df_test)
    
except Exception as e:
    print(f"❌ Connection failed: {e}")
```

```python
# Upload the DataFrame to the database
# If the table exists, it will be overwritten (if if_exists='replace')
# or appended (if if_exists='append')
# Define your table name here
try:
    df.to_sql(
        name=TABLE_NAME,          # Name of the table
        con=engine,               # The database engine connection
        if_exists='replace',      # Options: 'fail' (error), 'replace' (drop & recreate), 'append' (add to existing)
        index=False,              # Don't save the DataFrame's index (row numbers) as a column
        chunksize=1000,           # Optional: Write in chunks of 1000 rows for large datasets
        method='multi',           # Optional: Optimize for bulk insert
        dtype=None                # Optional: Define specific data types if needed
    )
    print(f"✅ Successfully uploaded {len(df)} rows to table '{TABLE_NAME}' in '{DB_NAME}'!")
except Exception as e:
    print(f"❌ Error during upload: {e}")
```
